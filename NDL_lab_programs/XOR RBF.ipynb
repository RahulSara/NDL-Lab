```python
import numpy as np
import matplotlib.pyplot as plt

# XOR dataset: Inputs and corresponding outputs
inputs = np.array([[0,0], [0,1], [1,0], [1,1]])
outputs = np.array([0, 1, 1, 0])  # Expected XOR outputs

# Define Gaussian Radial Basis Function (RBF)
def gaussian(vector, weights, sigma=0.25):
    return np.exp(-np.linalg.norm(vector - weights) ** 2 )

# Define RBF centers (Choosing XOR inputs as centers)
w1 = np.array([1,1])
w2 = np.array([0,0])

f1, f2 = [], []

print(f"Input\tFirst Function\tSecond Function")

# Compute RBF values for each input
for i in inputs:
    v1, v2 = gaussian(i, w1), gaussian(i, w2)
    f1.append(v1)
    f2.append(v2)
    print(f"{i} \t{v1:.4f} \t{v2:.4f}")


plt.scatter(f1,f2,c=outputs,cmap='coolwarm')
plt.plot([1,0],[0,1],'k--')
plt.xlabel('Hidden Func 1')
plt.ylabel('Hidden Func 2')
plt.show()
```
