```python
#universal approx

import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt

# Define a target function (Example: sin(2πx))
def target_function(x):
    return np.sin(2 * np.pi * x)

# Generate training data
x_train = np.linspace(0, 1, 100).reshape(-1, 1)  # Input: 100 points between 0 and 1
y_train = target_function(x_train)  # Compute target function values

# Build a simple feedforward neural network
model = keras.Sequential([
    keras.layers.Dense(10, activation='tanh', input_shape=(1,)),  # Hidden layer (UAT requirement)
    keras.layers.Dense(1)  # Output layer
])

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
model.fit(x_train, y_train, epochs=500, verbose=0)  # Train for 500 epochs

# Predict using the trained model
y_pred = model.predict(x_train)

# Plot the results
plt.figure(figsize=(6, 4))
plt.plot(x_train, y_train, label="Target Function (sin)")
plt.plot(x_train, y_pred, label="NN Approximation", linestyle="dashed")
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Universal Approximation of sin(2πx)")
plt.show()
```
