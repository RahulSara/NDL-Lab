```python
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def binary_step(x):
    return np.where(x >= 0, 1, 0)

def linear(x):
    return x

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def tanh(x):
    return np.tanh(x)

def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Subtract max(x) for numerical stability
    return exp_x / exp_x.sum(axis=0)

# Input data
x = np.linspace(-10, 10, 100)

# Plot activation functions
plt.figure(figsize=(10, 6))

plt.subplot(3, 3, 1)
plt.plot(x, binary_step(x))
plt.title("Binary Step")

plt.subplot(3, 3, 2)
plt.plot(x, linear(x))
plt.title("Linear")

plt.subplot(3, 3, 3)
plt.plot(x, sigmoid(x))
plt.title("Sigmoid")

plt.subplot(3, 3, 4)
plt.plot(x, relu(x))
plt.title("ReLU")

plt.subplot(3, 3, 5)
plt.plot(x, leaky_relu(x))
plt.title("Leaky_ReLU")

plt.subplot(3, 3, 6)
plt.plot(x, tanh(x))
plt.title("Tanh")

plt.subplot(3, 3, 7)
plt.plot(x, softmax(x))
plt.title("softmax")

plt.tight_layout()
plt.show()
```
